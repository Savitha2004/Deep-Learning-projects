```python
dataset=pd.read_csv('/content/drive/MyDrive/ML data set/melb_data.csv.zip')
dataset.head()
```

```python
dataset=dataset[:1000]
```

```python
dataset.nunique()
```

```python
dataset.shape
```

```python
dataset.columns
```

```python
cols_to_use=['Suburb', 'Address', 'Rooms', 'Type', 'Price', 'Method', 'SellerG',
       'Distance', 'Bedroom2', 'Bathroom', 'Car',
       'Landsize', 'BuildingArea', 'CouncilArea',
       'Regionname', 'Propertycount']
dataset=dataset[cols_to_use]
dataset.head()
```

```python
dataset.shape
```

```python
dataset.isna().sum()
```

```python
cols_to_fill_zero =['Propertycount','Distance','Bedroom2','Car']
dataset[cols_to_fill_zero]=dataset[cols_to_fill_zero].fillna(0)
```

```python
dataset.shape
```

```python
dataset.isna().sum()
```

```python
dataset['Landsize']= dataset['Landsize'].fillna(dataset.Landsize.mean())
dataset['BuildingArea']= dataset['BuildingArea'].fillna(dataset.BuildingArea.mean())
dataset.isna().sum()
```

```python
dataset.dropna(inplace=True)
dataset.isna().sum()
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset.head()
```

```python
x=dataset.drop('Price',axis=1)
y=dataset['Price']
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.3,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression().fit(train_x,train_y)
```

```python
print(reg.score(test_x,test_y))
print(reg.score(train_x,train_y))
```

```python
from sklearn import linear_model
lasso_reg=linear_model.Lasso(alpha=50,max_iter=100,tol=0.1)
lasso_reg.fit(test_x,test_y)
```

```python
lasso_reg.score(test_x,test_y)
lasso_reg.score(train_x,train_y)
```

```python
from sklearn.linear_model import Ridge
ridge_reg=Ridge(alpha=50,max_iter=100,tol=0.1)
ridge_reg.fit(test_x,test_y)
```

```python
print(ridge_reg.score(test_x,test_y))
print(ridge_reg.score(train_x,train_y))
```
