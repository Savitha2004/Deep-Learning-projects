```python
#Exp 2.1 Histogram
import cv2
import matplotlib.pyplot as plt
img=cv2.imread("/content/sq.jpg",0)
hist1=cv2.calcHist(img,[0],None,[256],[0,256])
img1=cv2.equalizeHist(img)
hist2=cv2.calcHist(img1,[0],None,[256],[0,256])
plt.subplot(221),plt.imshow(img),plt.title('original image')
plt.subplot(222),plt.plot(hist1),plt.title('histogram 1')
plt.subplot(223),plt.imshow(img1),plt.title('contrast image')
plt.subplot(224),plt.plot(hist2),plt.title('histogram 2')
```

```python
#exp 2.2 thresholding
import cv2
import matplotlib.pyplot as plt
img=cv2.imread("/content/sq.jpg",0)
_,t1=cv2.threshold(img,127,255,cv2.THRESH_BINARY)
_,t2=cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
_,t3=cv2.threshold(img,127,255,cv2.THRESH_TRUNC)
_,t4=cv2.threshold(img,127,255,cv2.THRESH_TOZERO)
_,t5=cv2.threshold(img,127,255,cv2.THRESH_TOZERO_INV)
plt.subplot(231),plt.imshow(img, cmap='gray'),plt.title('original image')
plt.subplot(232),plt.imshow(t1, cmap='gray'),plt.title('BINARY')
plt.subplot(233),plt.imshow(t2, cmap='gray'),plt.title('BINARY INVERSE')
plt.subplot(234),plt.imshow(t3, cmap='gray'),plt.title('TRUNCATE')
plt.subplot(235),plt.imshow(t4, cmap='gray'),plt.title('TOZERO')
plt.subplot(236),plt.imshow(t5, cmap='gray'),plt.title('TOZERO INVERSE')
```

```python
#exp-2.3 edge detection
import cv2
import matplotlib.pyplot as plt
img=cv2.imread("/content/sq.jpg",0)
edge=cv2.Canny(img,200,255)
plt.subplot(221),plt.imshow(img,cmap='gray'),plt.title('original image')
plt.subplot(222),plt.imshow(edge,cmap='gray'),plt.title('edge detection')
```

```python
#Exp 2.4 morphological analysis
import cv2
import matplotlib.pyplot as plt
import numpy as np
img=cv2.imread("/content/sq.jpg",0)
kernal=np.ones((5,5),np.uint8)
dia=cv2.dilate(img,kernal,iterations=1)
erosion=cv2.erode(img,kernal)
plt.subplot(221),plt.imshow(img),plt.title('original image')
plt.subplot(222),plt.imshow(dia),plt.title('Dilated image')
plt.subplot(223),plt.imshow(erosion),plt.title('Erosion image')
```
