```python
#import libraries
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
#load dataset
iris=load_iris()
#get features and target
x=iris.data
y=iris.target
```

```python
#preparing dataset to get dummy values
y=pd.get_dummies(y,dtype='int').values
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=20,random_state=4)
```

```python
print(x_train.size)
print(x_test.size)
print(y_train.size)
print(y_test.size)
```

```python
#Initialize hyperparameters and weights
#initialize valriables
learning_rate=0.1
iterations=5000
N=y_train.size
input_size=4
hidden_size=2
output_size=3
results=pd.DataFrame(columns=["mse","accuracy"])
print(results)
```

```python
np.random.seed(10)
w1=np.random.normal(scale=0.5,size=(input_size,hidden_size))
print(w1)
print()
w2=np.random.normal(scale=0.5,size=(hidden_size,output_size))
print(w2)
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```python
def mean_squared_error(y_pred,y_true):
  squared_error=(y_pred-y_true)**2
  sum_squared_error=squared_error.sum()
  mse=sum_squared_error/(2*y_pred.size)
  return mse
```

```python
def accuracy(y_pred,y_true):
  acc=y_pred.argmax(axis=1)==y_true.argmax(axis=1)
  return acc.mean()
```

```python
Z1=np.dot(x_test,w1)
A1=sigmoid(Z1)

Z2=np.dot(A1,w2)
A2=sigmoid(Z2)

acc=accuracy(A2,y_test)
print("accuracy: ",acc)
```

```python
for i in range(iterations):
  Z1=np.dot(x_train,w1)
  A1=sigmoid(Z1)
  Z2=np.dot(A1,w2)
  A2=sigmoid(Z2)
  mse=mean_squared_error(A2,y_train)
  acc=accuracy(A2,y_train)
  results=pd.concat([results,pd.DataFrame({"mse":mse,"accuracy":acc},index=[0])],ignore_index=True)
  print(results)
  E1=A2-y_train
  dw1=E1*A2*(1-A2)
  print("dw1")
  print(dw1)
  print(dw1.size)
  print()
  E2=np.dot(dw1,w2.T)
  dw2=E2*A1*(1-A1)
  print("dw2")
  print(dw2)
  print(dw2.size)
  print()

  #weight updates
  w2_update=np.dot(A1.T,dw1)/N
  w1_update=np.dot(x_train.T,dw2)/N

  w2=w2-learning_rate*w2_update
  w1=w1-learning_rate*w1_update
  print()
  print('w2_update',w2_update)
  print('w1_update',w1_update)
  print('w2',w2)
  print('w1',w1)


```

```python
results.mse.plot(title="Mean Squared Error")
```

```python
results.accuracy.plot(title="Accuracy")
```

```python
#feed forward
#Hidden layer
Z1=np.dot(x_test,w1)
A1=sigmoid(Z1)
Z2=np.dot(A1,w2)
A2=sigmoid(Z2)
acc=accuracy(A2,y_test)
print("Accuracy:{}".format(acc))
```
